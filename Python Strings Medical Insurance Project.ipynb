```python
#This code has the objetive to pratice Strings functions in Phyton.

medical_data = \
"""Marina Allison   ,27   ,   31.1 , 
#7010.0   ;Markus Valdez   ,   30, 
22.4,   #4050.0 ;Connie Ballard ,43 
,   25.3 , #12060.0 ;Darnell Weber   
,   35   , 20.6   , #7500.0;
Sylvie Charles   ,22, 22.1 
,#3022.0   ;   Vinay Padilla,24,   
26.9 ,#4620.0 ;Meredith Santiago, 51   , 
29.3 ,#16330.0;   Andre Mccarty, 
19,22.7 , #2900.0 ; 
Lorena Hodson ,65, 33.1 , #19370.0; 
Isaac Vu ,34, 24.8,   #7045.0"""

# Add your code here
#Replace # to $
update_medical_data= medical_data.replace("#", "$")
print(update_medical_data)

#Create a variable with the number of records
num_records=0
for each in update_medical_data:
   if each == "$":
    num_records +=1
print("\nThere are {} medical records in the data".format(num_records))

#Another way using List Comprehension
#num_records= [ each for each in update_medical_data if each == "$"]
#print(len(num_records))

#Splitting Strings
medical_data_split = update_medical_data.split(";")
print (medical_data_split)

medical_records=[]
for each in medical_data_split:
  medical_records.append(each.split(","))

print("\n")
print(medical_records)

medical_records_clean=[]
for record in medical_records:
  record_clean=[]
  for item in record:
    record_clean.append(item.strip())
  medical_records_clean.append(record_clean)
print("\n")
print (medical_records_clean)

#Analyzing Data 
#1st print patient names
print("\nPatien Names")
for record in medical_records_clean:
  print(record[0].upper())

#We will storage each name, age, BMI and insurance cost at their own list

names=[]
ages= []
bmis= []
insurance_costs= []

for each in medical_records_clean:
  names.append(each[0])
  ages.append(each[1])
  bmis.append(each[2])
  insurance_costs.append(each[3])
print("\nPatient Names")
print (names)
print("\nPatient Ages")
print(ages)
print("\nPatient BMI's")
print(bmis)
print("\nPatiente Insurance Costs")
print(insurance_costs)

#Calculate BMI Average

total_bmi= 0
for each in bmis:
  total_bmi += float(each)  
print("\nTotal BMI is {}.".format(total_bmi))
average_bmi = total_bmi / len(bmis)
print("\nAverage BMI: {}".format(average_bmi))

#Extra coding
#Calculate average of insurance costs, for that we would have to remove the $
insurance_math=[]
for each in insurance_costs:
  insurance_math.append(each.replace("$",""))
print(insurance_math)
#Calculate the average for insurance
insurance_total =0
for cost in insurance_math:
  insurance_total += float(cost)
#print("\nThe total cost of insurance is: "+str(insurance_total))

averaga_insurance_cost = insurance_total / len(insurance_math)
print("\nThe average insurance cost is {}".format(averaga_insurance_cost))

#Print loops
print("\nBelow it sh")
for i in range(len(names)):
  #for age in ages:
    #for bmi in bmis:
      print ("{name} is {age} years old with a BMI of {bmi} and an insurance cost of {cost}.".format(name=names[i], age=ages[i], bmi= bmis[i],cost=insurance_costs[i] ))












```
